```python
from transformers import AutoTokenizer
from datasets import load_dataset
```

```python
tokenize = AutoTokenizer.from_pretrained("bert-base-uncased")
imdb_data = load_dataset("imdb")
imdb_data
```

### Preprocess the data , Convert text to Vectors

```python
def text_tokenizer(examples):
    return tokenize(examples['text'],padding='max_length',truncation=True)

tokenized_dataset = imdb_data.map(text_tokenizer, batched=True)
tokenized_dataset
```
